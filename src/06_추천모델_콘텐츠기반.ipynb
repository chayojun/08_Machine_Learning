```python
import pandas as pd

df = pd.read_csv("../data/tmdb_5000_movies.csv")
df.head()
```

```python
movies_df = df[["id", "title", "genres", "vote_average", "vote_count", "popularity", "keywords", "overview"]]

pd.set_option("max_colwidth", 100)
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df.iloc[0]["genres"]
```

```python
from ast import literal_eval

result = literal_eval(movies_df.iloc[0]["genres"])
result
```

```python
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['genres']
```

```python
movies_df.iloc[0]["keywords"]
```

```python
movies_df["keywords"] = df["keywords"].apply(literal_eval)
movies_df["keywords"]
```

```python
name_result = []
for genre_temp in movies_df["genres"]:
    names = []
    for temp in genre_temp:
        name = temp["name"]
        names.append(name)

    name_result.append(names)
    

movies_df["genres"] = name_result
```

```python
movies_df["genres"]
```

```python
data_result = []
for temp in movies_df["genres"]:
    data = (" ").join(temp)
    data_result.append(data)

movies_df["genres_literal"] = data_result
movies_df["genres_literal"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df = 1, ngram_range = (1, 2))
genre_vect = count_vect.fit_transform(movies_df["genres_literal"])
count_vect.vocabulary_
```

```python
genre_vect.shape # 4803개의 영화, 276개의 장르
```

```python
genre_vect_array = genre_vect.toarray()
genre_vect_array
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_vect_array, genre_vect_array)
genre_sim[0].argsort()[::-1]
```

## 컨텐츠 기반의 추천 시스템

```python
def find_sim_movie(df, sorted_index, title_name, top_n=10):
    
    title_movie =  df[df["title"] == title_name]
    title_index = title_movie.index.values

    similar_index = sorted_index[title_index, : (top_n)]
    similar_index = similar_index.reshape(-1)

    return df.iloc[similar_index]

genre_sim__sorted = genre_sim.argsort()[:,::-1]

movies_result = find_sim_movie(df, genre_sim__sorted, "Avatar", 10)
movies_result[["title"]]
```

```python
keyword_result = []
for keywords_temp in movies_df["keywords"]:
    keywords = []
    for temp in keywords_temp:
        keyword = temp["name"]
        keywords.append(keyword)
    
    keyword_result.append(keywords)

movies_df["keywords"] = keyword_result
```

```python
data_result = []
for temp in movies_df["keywords"]:
    data = (" ").join(temp)
    data_result.append(data)

movies_df["keywords_literal"] = data_result
movies_df["keywords_literal"]
```

```python
count_vect = CountVectorizer(min_df = 1, ngram_range = (1, 2))
keywords_vect = count_vect.fit_transform(movies_df["keywords_literal"])
count_vect.vocabulary_
```

```python
keywords_vect.shape
```

```python
keywords_vect_array = keywords_vect.toarray()
keywords_vect_array
```

```python
keywords_sim = cosine_similarity(keywords_vect_array, keywords_vect_array)
keywords_sim[0].argsort()[::-1]
```
